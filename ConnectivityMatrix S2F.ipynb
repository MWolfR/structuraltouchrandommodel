```python
import numpy
import pandas

import conntility

fn_mat = "/gpfs/bbp.cscs.ch/project/proj159/home/reimann/human_randomized_touch_matrices.h5"
prefix="randomize_edges_and_count"
group_name="instance0"

fn_actual_mat = "/gpfs/bbp.cscs.ch/project/proj159/home/barros/conn_matrix/Human_960um_squared_funct_conmat.h5"
fn_s2f_params = "/gpfs/bbp.cscs.ch/project/proj159/circuit_building/270923/run/connectome/hncx_neurons__hncx_neurons_chemical_synapse/functional/spykfunc/_debug/01_pathway_params.csv"

M = conntility.ConnectivityMatrix.from_h5(fn_mat, prefix=prefix, group_name=group_name)
s2f_params = pandas.read_csv(fn_s2f_params)
```

## Read s2f parameters

```python
s2f_params = pandas.read_csv(fn_s2f_params)

def parse_pathway_str(str_in):
    k, v = zip(*[_x.split("=") for _x in str_in.split()])
    return pandas.Series(v, index=k)

s2f_params = pandas.concat([s2f_params, s2f_params["pathway_str"].apply(parse_pathway_str)], axis=1)
s2f_params
```

## Stage 1: random reduction

```python
from scipy.stats import binom

edge_mtypes = M.edge_associated_vertex_properties("mtype")
for _, row in s2f_params.iterrows():
    row = row.rename({"src_mtype_i": "row", "dst_mtype_i": "col"})
    mask = edge_mtypes == row[["row", "col"]]
    mask = mask.all(axis=1)
    print(mask.sum())
    if mask.sum() == 0:
        continue
    M._edges["count"][mask] = binom(M._edges["count"][mask], row["pP_A"]).rvs()

M = M.filter_with_function(lambda df: df["count"] > 0)
```

## Stage 2: Cut!

```python
from scipy.stats import norm

def cut_filter(mn, cv):
    sigma = mn * cv
    distr = norm(mn, sigma)
    def ret_fun(edge_table):
        p = distr.cdf(edge_table["count"])
        return numpy.random.rand(len(edge_table)) < p
    return ret_fun


edge_mtypes = M.edge_associated_vertex_properties("mtype")
for _, row in s2f_params.iterrows():
    row = row.rename({"src_mtype_i": "row", "dst_mtype_i": "col"})
    mask = edge_mtypes == row[["row", "col"]]
    mask = mask.all(axis=1)
    print(mask.sum())
    if mask.sum() == 0:
        continue
    M = M.filter_with_function(cut_filter(row["pMu_A"], 0.33), mask=mask)
    edge_mtypes = M.edge_associated_vertex_properties("mtype")
```

## Stage 3: Randomly remove connections until we match the reference

```python
actual_M = conntility.ConnectivityMatrix.from_h5(fn_actual_mat)

assert len(actual_M) == len(M)
reference_counts = actual_M.edge_associated_vertex_properties("mtype").value_counts()
```

Two variants:

One, match the overall number of edges to the (functional) reference.
Two, match the number of edges in each pathway to the (functional) reference.

Version two is possibly not supported, because the control already has fewer edges than the reference in some pathways.

```python
version = 1
```

```python
def randomly_pick_number(N):
    def ret_fun(df):
        assert len(df) >= N
        rnd = numpy.hstack([
            numpy.ones(N, dtype=bool),
            numpy.zeros(len(df) - N, dtype=bool)
        ])
        return numpy.random.permutation(rnd)
    return ret_fun

print("Before:\nControl has {0} edges.\nReference has {1} edges".format(
    len(M.edges), reference_counts.sum()
))

if version == 1:
    M = M.filter_with_function(randomly_pick_number(reference_counts.sum()))
elif version == 2:
    edge_mtypes = M.edge_associated_vertex_properties("mtype")
    for types, N in reference_counts.items():
        mask = edge_mtypes == types
        mask = mask.all(axis=1)
        print(mask.sum())
        if mask.sum() == 0:
            continue
        M = M.filter_with_function(randomly_pick_number(N), mask=mask)
        edge_mtypes = M.edge_associated_vertex_properties("mtype")

print("After:\nControl has {0} edges.\nReference still has {1} edges".format(
    len(M.edges), reference_counts.sum()
))
```

```python
from connalysis.modelling import conn_prob_2nd_order_model

p_ctrl = conn_prob_2nd_order_model(M.matrix.tocsc().astype(bool), M.vertices)
p_ref = conn_prob_2nd_order_model(actual_M.matrix.tocsc().astype(bool), actual_M.vertices)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure()
ax = fig.gca()
x = numpy.linspace(1, 800)

A_ctrl = p_ctrl["exp_model_scale"].iloc[0]
B_ctrl = p_ctrl["exp_model_exponent"].iloc[0]
con_prob_ctrl = A_ctrl * numpy.exp(-x * B_ctrl)

A_ref = p_ref["exp_model_scale"].iloc[0]
B_ref = p_ref["exp_model_exponent"].iloc[0]
con_prob_ref = A_ref * numpy.exp(-x * B_ref)

plt.plot(x, con_prob_ref)
plt.plot(x, con_prob_ctrl)
#plt.gca().set_ylim([0, 1])
#plt.gca().set_yscale("log")
```
